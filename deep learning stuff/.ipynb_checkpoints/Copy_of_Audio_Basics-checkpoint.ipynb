Audio data analysis is about analyzing and understanding audio signals captured by digital devices, with numerous applications in the enterprise, healthcare, productivity, and smart cities. Applications include customer satisfaction analysis from customer support calls, media content analysis and retrieval, medical diagnostic aids and patient monitoring, assistive technologies for people with hearing impairments, and audio analysis for public safety. Other applications include speech recognition, trigger word detection, environmental noise analysis and bioacoustics.

The sound excerpts are digital audio files in .wav format. Sound waves are digitized by sampling them at discrete intervals known as the sampling rate (typically 44.1kHz for CD-quality audio meaning samples are taken 44,100 times per second). This means that in one second of data, you will have 44,100 data points. Similarly for different sampling rates.

Each sample is the amplitude of the wave at a particular time interva

In signal processing, sampling is the reduction of a continuous signal into a series of discrete values. The sampling frequency or rate is the number of samples taken over some fixed amount of time. A high sampling frequency results in less information loss but higher computational expense, and low sampling frequencies have higher information loss but are fast and cheap to compute

Adult humans fundamental voice frequency range is between 85Hz to 255Hz (85Hz to 180Hz for male and 165Hz to 255Hz for female). On top of the fundamental frequency there are harmonics of fundamental frequencys. Harmonics are whole multiplications of the fundamental frequency. If for instance the fundamental frequency is 100Hz then its second harmonic will be 200Hz, third harmonic is 300Hz and so on.

You can see an example in the image below which shows frequency vs. time of several pronounced words and color represents frequency power at that point (yellow strongest and black weakest). A spectrogram is a visual way of representing the signal strength, or “loudness”, of a signal over time at various frequencies present in a particular waveform.  Not only can one see whether there is more or less energy at, for example, 2 Hz vs 10 Hz, but one can also see how energy levels vary over time.  



![](https://static.scientificamerican.com/sciam/assets/media/sound/speechsep-audio/speechsep-2-spect.png)


Notice the first horizontal yellow line on the bottom of each segment. That is the fundamental frequency and its the strongest. Above that there are harmonics.

Spectrograms are basically two-dimensional graphs, with a third dimension represented by colors. Time runs from left (oldest) to right (youngest) along the horizontal axis. The amplitude (or energy or “loudness”) of a particular frequency at a particular time is represented by the third dimension, color, with dark blues corresponding to low amplitudes and brighter colors up through red corresponding to progressively stronger (or louder) amplitudes.

Humans can hear roughly between 20Hz to 20KHz. The perception of sound is non-linear and you can better distinguish between low frequency sounds than high frequency sounds e.g. humans can clearly hear the difference betwee 100Hz and 200Hz but not between 15kHz and 15.1kHz.

You can try in usin a tone generator: http://www.szynalski.com/tone-generator/

A MEL scale is a unit of PITCH proposed by Stevens, Volkmann and Newmann in 1937. The MEL scale is a scale of pitches judged by listeners to be equal in distance one from another. Because of how humans perceive sound the MEL scale is a non-lenear scale and the distances between the pitches increeses with frequency.

![](https://upload.wikimedia.org/wikipedia/commons/thumb/a/aa/Mel-Hz_plot.svg/450px-Mel-Hz_plot.svg.png)

Acording to Nyquist, that sampling rate of 44.1kHz give us band-width of 22.05kHz. The 0Hz to 22.05kHz zone is called the first Nyquist zone, 22.05kHz to 44.1kHz is the second Nyquist zone and so on. If there is a signal in the second Nyquist zone (lets say 22.1kHz) it will alias to the first Nyquist zone (22kHz in our case). Armed with this knowledge, you'll know what sampling rate to record at when collecting data.







To generate a spectrogram, a time-domain signal is divided into shorter segments of equal length. Then, the fast Fourier transform (FFT) is applied to each segment. The spectrogram is a plot of the spectrum on each segment.

## Read matlab values

```python
mat = scipy.io.loadmat('md15_156a_regclicks_dive1.mat')
```

```python
mat.get('CL1').shape
```

## Get the annotated values

```python
vals = np.asarray(mat.get('CL1'))
```

```python
vals[0:10]
```

## Extract a first segment
Say we want to extract a 2 second segment starting from the first value but going a bit "to the left"

Set starting values. Round down

```python
start_sec = math.floor(vals[0])
start_sec
```

Set ending value. Round up

```python
end_sec = math.floor(vals[0])+2
end_sec
```

## Extract the values from the waveform

```python
segment = audio_amps[start_sec*audio_sample_rate:end_sec*audio_sample_rate]
```

## Check length

```python
len(segment)
```

## Check segment duration

```python
len(segment)/audio_sample_rate
```

## Generate spectrogram for segment

```python
S = librosa.feature.melspectrogram(y=segment, 
                                   sr=audio_sample_rate,
                                   n_mels=128,
                                   n_fft=4096, 
                                   hop_length=2048)

S_dB = librosa.power_to_db(S, ref=np.max)
S_dB_copy = librosa.power_to_db(S, ref=np.max)
```

```python
librosa.display.specshow(S_dB, 
                         sr=audio_sample_rate, 
                         hop_length=2048, 
                         x_axis='time', y_axis='mel')
```

## Display shape of spetrogram data

```python
S_dB.shape
```

## How did we get that shape exactly? This is how we can calculate it...

sample rate divided by the hop length. Then multiply by the number of seconds. Take the ceiling (round up) due to potential padding. This matches the 235 above. So this is how to determine positions within the features and how to map seconds to the feature space.



```python
math.ceil((audio_sample_rate/2048)*2)
```

# Example 1

```python
S_dB = librosa.power_to_db(S, ref=np.max)
```

## Get the first annotated point and determine how many seconds it is from the start of the segment

```python
position = vals[0][0] - start_sec
position
```

Here, determine it's corresponding position within the feature space defined above (length 235)

```python
math.ceil(audio_sample_rate/2048*position)
```

This means that 0.307 seconds corresponds to position 36!

Below the spectrogram with a line draw at time = 0.307 in green

```python
librosa.display.specshow(S_dB, sr=audio_sample_rate, hop_length=2048, x_axis='time', y_axis='mel')
plt.vlines(position, 0, 70000, color='g')
```

Below, the same idea, except I'm not using time, I'm setting all values in position 36 to be 1, which will generate a line. This shows that we can have a direct mapping between time and the fourier space. The lines above and below match perfectly.

```python
S_dB.T[36] = 1
librosa.display.specshow(S_dB, sr=audio_sample_rate, hop_length=2048, x_axis='time', y_axis='mel')
```

## MFCCs

I won't go into the details here and leave it up to you to learn about Mel-frequency cepstral coefficients. But they are somtimes used in speech related recognition tasks. However, in modern literature they aren't used as often, at least with respect to animal vocalisations.

```python
S = librosa.feature.melspectrogram(y=segment, sr=audio_sample_rate, n_mels=128)
mfcc = librosa.feature.mfcc(S=librosa.power_to_db(S))
```

```python
mfcc.shape
```

```python
librosa.display.specshow(mfcc, sr=audio_sample_rate, x_axis='time')
```

## Another example

```python
y, sr = librosa.load(librosa.ex('trumpet'))
```

```python
librosa.display.waveplot(y, sr=sr)
```

```python
S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
mfcc = librosa.feature.mfcc(y, n_mfcc=40, sr=sr)
```

```python
librosa.display.specshow(mfcc, sr=sr, x_axis='time')
```

```python
fig, ax = plt.subplots()
S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128,
                                    fmax=8000)
S_dB = librosa.power_to_db(S, ref=np.max)

img = librosa.display.specshow(S_dB, x_axis='time',
                         y_axis='mel', sr=sr,
                         fmax=8000, ax=ax)
fig.colorbar(img, ax=ax, format='%+2.0f dB')
ax.set(title='Mel-frequency spectrogram')
```

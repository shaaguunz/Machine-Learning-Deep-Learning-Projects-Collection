```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url='https://www.imdb.com/search/title/?title_type=tv_series&num_votes=,100000&sort=runtime,asc'
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
Title=[]
premiere_year=[]
rating=[]
time=[]
genres=[]
stars=[]
plots=[]
```

```python
series_container= soup.find_all('div', class_ = 'lister-item mode-advanced')
```

```python
for series in series_container:
        
        title=series.h3.a.get_text()
        Title.append(title)
        year=series.find('span', class_ = 'lister-item-year text-muted unbold').get_text()
        premiere_year.append(year[1:5])        
        rate=float(series.strong.text)
        rating.append(rate)
        runtime = series.p.find('span', class_ = 'runtime').get_text()
        time.append(runtime)
        
        if series.p.find('span', class_ = 'genre') is not None:
            genre = series.p.find('span', class_ = 'genre').text.replace("\n", "").rstrip()
            genres.append(genre)
        else:
            genres.append("")
            
            
        
        
        if series.find("p",class_="text-muted") is not None:
            plot=  series.find("p",class_="text-muted")
            plots.append(plot.find_next('p').text.replace("\n",""))
        
        else:
            plots.append("")
        if series.find("p",class_="text-muted") is not None:
            st=  series.find("p",class_="text-muted")
            stars.append(st.find_next('p').find_next('p').text.replace(":"," ").replace("\n","").lstrip()[5:])
        
        else:
            stars.append("")
        
        
        
```

```python
df = pd.DataFrame({
    "title": Title,
    "premiere year": premiere_year,
    "rating": rating,
    "time":time,
    "genre":genres,
    "star":stars,
    "plot":plots
    
})
```

```python
df.to_csv("df.csv")
```
